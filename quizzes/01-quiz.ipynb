### (a) Transform a list of values and plot it

### (c) Plot random variables

- Let $n = 1000$
- Generate `vec_x` from a Normal with (loc = 2, scale = 1, size = n)
- Generate `vec_z` from a chi square with (df = 2, size = n)
- Generate `vex_y` from a uniform with  (low = 5,high = 10, size = n)
- Plot two scatter plots: (`vec_x` vs `vec_y`) and (`vec_x` vs `vec_z`)
    - Label the axes and title
    - Each graph should have a different title
 
- To get full points, plot the scatter plots on a single row using the `subplots` syntax (see Lecture 6)

```python
# Write your answer here:

n = 1000

vec_x = np.random.normal (loc = 2, scale = 1, size = n)
vec_z = np.random.chisquare (df = 2, size = n)
vec_y = np.random.uniform (low = 5,high = 10, size = n)

fig, lsit_subfig = plt.subplots(1,2, figsize=(10,3))
plt.tight_layout()
plt.subplots_adjust(wspace=0.3)

#first figure
lsit_subfig[0].scatter(x=vec_y, y=vec_x)
lsit_subfig[0].set_title("uniform vs normal distribution")
lsit_subfig[0].set_xlabel("uniform distribution")
lsit_subfig[0].set_ylabel("normal distribution")

#second figure
lsit_subfig[1].scatter(x=vec_z, y=vec_x)
lsit_subfig[1].set_title("chi square and normal distribution")
lsit_subfig[1].set_xlabel("chi square")
lsit_subfig[1].set_ylabel("normal distribution")


```

### (d) Create the following list of months:

` list_months = ["August","February"] ` 

` list_spring_semester = ["January", "February", "March", "April", "May"] `

Create a for-loop that checks whether each element of `list_months` belongs to the spring semester. Create a message with this result. 

The output generated by your loop should look like this:

`"August is NOT part of the Spring semester"`

`"February is part of the Spring semester"`

- **Hint**: Use if/else statements inside the loop. You can use the command "in" (see Lecture 4) to test whether an element belongs to a list


```python
# Write your answer here:


list_months = ["August","February"]
list_spring_semester = ["January", "February", "March", "April", "May"]


for month in list_months:
    if month in list_spring_semester:
        print(f"{month} is part of the Spring semester")
    else:
        print(f"{month} is NOT part of the Spring semester")

```
